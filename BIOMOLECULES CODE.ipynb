```python
#BIOMOLECULES CODE
#Glycine 

import rdkit 
from rdkit import Chem
from rdkit.Chem import Draw
from rdkit.Chem.Draw import IPythonConsole
from rdkit.Chem import Descriptors 
from rdkit.Chem import AllChem
from rdkit import DataStructs
from rdkit import Chem
from rdkit.Chem import Crippen
from rdkit.Chem.EnumerateStereoisomers import EnumerateStereoisomers, StereoEnumerationOptions

IPythonConsole.drawOptions.addAtomIndices = True
IPythonConsole.molSize = 400,400

```

```python
mol = Chem.MolFromSmiles("C(C(=O)O)N")
mol
```

```python
mw = Descriptors.ExactMolWt(mol)
print ("Molecular wt = ", mw)

TSA = Descriptors.TPSA(mol)
print ("Topological Surface area = ", TSA, "Å²")

hacc = Descriptors.NumHAcceptors(mol)
print ("Number of Hydrogen bond Acceptors = ", hacc)

hdon = Descriptors.NumHAcceptors(mol)
print ("Number of Hydrogen bond donors = ", hdon)

rb = Descriptors.NumRotatableBonds(mol)
print ("number of rotatable bonds = ", rb)

logP = Crippen.MolLogP(mol)
print ("ClogP = ", logP)

HA = Descriptors.NumHeteroatoms(mol)
print ("Heteroatoms = ", HA)

hac = Descriptors.HeavyAtomCount(mol)
print ("Number of Heavy atoms =", hac)

ah = Descriptors.NumAromaticHeterocycles(mol)
print ("Number of Aromatic Heterocycles =", ah)

print("SP3 hybridization count =", sum((x.GetHybridization() == Chem.HybridizationType.SP3) for x in mol.GetAtoms()))

SP = Descriptors.FractionCSP3(mol)
print ("Fraction of SP3 Carbons =", SP)

print("Chiral centers = ", Chem.FindMolChiralCenters(mol,force=True,includeUnassigned=True,useLegacyImplementation=False))

m = Chem.MolFromSmiles('C(C(=O)O)N')
isomers = tuple(EnumerateStereoisomers(m))
len(isomers)
16
for smi in sorted(Chem.MolToSmiles(x, isomericSmiles=True) for x in isomers):
    print("stereoisomers -", smi)




print ("ONLY ACHIRAL AMINO ACID")
```
